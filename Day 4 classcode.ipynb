```python
class member:
    def __init__(self):
        print(" New member has been added")
        
member1 = member()
member2 = member()
member3 = member()
```

```python
class member:
    def __init__(self):
        self.name = "john"

member1 = member()
member2 = member()
print(member1.name)
print(member2.name)
```

```python
class member:
    def __init__(self , first_name):
        self.name = first_name 

member1 = member("mark")
member2 = member("sally")
print(member1.name)
print(member2.name)
```
